## PC COMPONENTES

```python
#SS
url = 'https://www.pccomponentes.com/televisores/3840-x-2160-4k-uhd/60-pulgadas/65-pulgadas/led/samsung'
#LG
#url = 'https://www.pccomponentes.com/televisores/3840-x-2160-4k-uhd/60-pulgadas/65-pulgadas/led/lg'
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
table = soup.find_all('div', {'class': ['col-xs-12 col-xl-9']})[0]
table
```

```python
rows = table.find_all(['div', 'a', 'header', 'h3'], {'class' : ['c-product-card__header', 'c-product-card__prices cy-product-price']})
rows_parsed = [row.text for row in rows]
rows_parsed
```

```python
row_split = 2
rows_refactored = [rows_parsed[x:x+row_split] for x in range(0, len(rows_parsed), row_split)]
rows_refactored

```

```python
df = pd.DataFrame(rows_refactored, columns=['Model', 'Price'])
df
```
